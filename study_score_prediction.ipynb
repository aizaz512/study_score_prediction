```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel('/content/linear dataset.xlsx')
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Study_Hours'], df['Exam_Score'])
plt.xlabel('Study_Hours)', color = "red")
plt.ylabel('Exam_Score', color = "red")
plt.title('Study_Hours')
plt.show()
```

```python
lr = linear_model.LinearRegression()
lr.fit(df[['Study_Hours']], df.Exam_Score)
```

```python
lr.predict([[6]])
```

```python
lr.score(df[['Study_Hours']], df.Exam_Score)
lr.coef_
```

```python
lr.intercept_
```

```python
predicted = 5.71678322 * 6 + 31.424242424242422
print(predicted)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Study_Hours'], df['Exam_Score'])
plt.xlabel('Study_Hours', color = "red")
plt.ylabel('Exam_Score', color = "red")
plt.title('Study_Hours vs. Exam_Score')
plt.plot(df['Study_Hours'], lr.predict(df[['Study_Hours']]), color = 'red')
plt.show()
```
